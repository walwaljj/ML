# 인공지능/머신러닝/딥러닝

* 인공지능(Artificail Intelligence): 사람처럼 똑똑한 기계 (가장 상위단계)
* 머신러닝(Machine Learning): 인공지능을 구현하는데 성공한 방법
* 딥러닝(Deep Learning): 머신러닝보다 더 똑똑한 인공지능을 구현하는 데 성공한 방법

# 전통적인 프로그래밍 VS 머신러닝의 차이점

* 프로그래밍 방식: 공식을 알려주고 답을 구하게 한 것
    * 규칙을 세우고 이를 프로그램으로 구현한 후, 데이터를 입력하여 해답을 찾음
    * 변화가 생기면 다시 규칙을 세워야 해서 변화 대응이 어려움
* 기계학습 방식(머신러닝): 답을 알려주고 공식을 깨우치게 한 것
    * 어렸을 때 한글 익히는 방식과 동일 
    * 강아지 사진 또는 실물은 데이터(data)이며, '강아지'라는 이름은 레이블(label)임.

사람은 데이터와 해답을 알려줌
> 기계는 데이터를 학습함.

# 기계학습 방법

* 머신러닝 방법은 알고리즘과 데이터를 입력하는 형태에 따라 구분합니다.
* 레이블(정답)의 유무에 따라
* 레이블이 있으면 **지도 학습**
* 레이블이 없으면,**비지도 학습**
* 행동에 따라 보상하여 인공지능을 향상시키는 **강화학습**이 있습니다.


* **지도 학습** : 문제와 답을 가르쳐 주면서 공부시키는 방식(친절함)
* **비지도 학습**:문제만 잔뜩 주고 알아서 답을 찾으라고 하는 방식(불 친절함)
* **강화 학습**: 문제도 답도 아무것도 안 주고, 그냥 컴퓨터를 어떤 장소(환경)에 집어 넣고, 그러면 컴퓨터는 자기 마음대로 행동하게 되는데, 그렇게 행동을 할 때마다 상점 또는 벌점을 줘서 채점을 합니다. 그러면 컴퓨터는 점점 상점을 많이 받는 방식으로 스스로 진화하는 방식

# 기계학습 모델

* 지도학습 모델에는 **예측(회귀)** 과 **분류(Classification)** 가 있고,
* 비지도 학습 모델에는 **군집(Clustering)** 이 있습니다. 

## 예측(회귀)

* 주어진 입력과 출력의 쌍을 학습한 후,
* 새로운 데이터를 입력했을 때,
* 합리적인 출력값을 예측하는 것입니다. 
* 훈련 데이터는 **연속적인 값**으로 입력과 출력이 모두 실수입니다. 
* 예)김장철 배추 가격 예측, 상품 판매량 예측, 날씨에 대한 예측, 주택 가격의 예측

image.png

## 분류
* 훈련 데이터에 각각 레이블(클레스라고 얘기함.)을 지정하여 학습하여 분류합니다. 
* 예측에서는 연속적인 값을 출력했지만,
* 분류에서는 연속되지 않고 **단절되어 있는 값**을 출력합니다. 
* 개 사진을 보고 개인지 아닌지 분류하면 이항 분류(2가지)
* 개,고양이 토끼 사진을 분류하면 다항 분류(3가지부터)
* 예) 스팸메일 분류, 손 글씨 분류, 암환자 분류, 0~9까지의 숫자 구분

# 데이터 속성(featuer)과 유형

* 데이터 속성은 다른 것과 구별할 수 있는 성질을 말합니다. 
* 데이터는 관찰이나 실험을 통해 수집되는 값(관측치) 또는 특성을 의미하며, 하나의 데이터는 여러가지 ***속성*** 들로 구성됩니다. 
* 붓꽃 데이터 속성(Feature)
    * Sepal_length(꽃받침 길이)
    * Sepal_width(꽃받침 너비)
    * Petal_length(꽃잎 길이)
    * Petal_width(꽃잎 너비)

* 붓꽃 품종은 (Class)
    * Setosa
    * Versicolor
    * Virgnica

    

한 행을 관측치 = 인스탠스 + 인스탠스 + ... => 샘플



> 넘파이는 차원 , 디맨션
> 클래스 레이블 = 타겟


* 데이터 자료형

    * 연속형 데이터 : 꽃잎의 길이와 너비는 실수 형태의 값을 가지는 **연속형 데이터** 입니다. (회기)
    * 범주형 데이터(카테고리 데이터) : 꽃의 품종은 문자 형태의 정해진 값을 가지는 **범주형 데이터** 입니다. (분류 .. 바차트)  

* Feature와 Class의 관계를 이야기 할 때

    * 붓꽃의 꽃받침 길이와 너비, 꽃잎의 길이와 너비가 품종에 어떤 영향을 미치는지 관계를 파악할 때,
    * 결과에 영향을 미치는 원인이 되는 변수를 **독립변수** 라고 합니다. (Sepal_length, Sepal_width, Petal_length, Petal_width) 
    * 특성값에 따라 결정되는 변수를 **종속변수** 라고 합니다.  (품종)

## 데이터 속성

* 피처(Feature)
    * 피처는 데이터 세트의 일반 속성입니다. 
    * 레이블을 제외한 나머지 **속성**을 모두 피처로 지칭합니다. 
    
* 레이블(Label)
    * 데이터의 학습을 위해 주어진 **정답** 데이터입니다.

## 언더 피팅 VS 오버 피팅

* 머신러닝의 최종 목표는 **일반화**에서도 좋은 성능을 내는 것입니다. 
    * 고3이 공부하는 이유는 모의고사 만점이 아니라, 
    * 어떤 문제가 출제될 지 모르는 수능에서의 만점이 목표입니다.

* **언더 피팅(Underfitting)**
    * 학습이 부족한 상태
    * 지나친 단순화로 인해 에러가 많이 발생한 상태

* **오버 피팅(Overfitting)**
    * 학습이 너무 많이 된 상태
    * 학습 데이터에 대해 과하게 학습되어, 학습 데이터 이외의 데이터에 대해서는 잘 동작하는 못하는 상태
