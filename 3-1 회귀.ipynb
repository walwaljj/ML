## 회귀 개요

* 회귀는 현대 통계학을 이루는 큰 축입니다. 
* 그런데, 회귀라는 말 자체가 친근하지 않은 이상하게 들리는 말입니다. 도대체 뭘 회귀한다는 건지...


* **"부모의 키가 크더라도 자식의 키가 대를 이어 무한정 커지지 않으며, 부모의 키가 작더라도 대를 이어 자식의 키가 무한정 작아지지 않는다"**
* 회귀 분석은 이처럼 데이터 값이 **평균**과 같은 일정한 값으로 돌아가려는 경향을 이용한 통계학 기법입니다. 


* 회귀는 여러개의 피처(독립변수)와 한 개의 레이블(종속변수) 간의 상관관계를 모델링하는 기법입니다. 
* 연속형 값을 예측하기 위한 모델

## 모델 적합 과정

* 주어진 데이터를 가장 잘 설명할 수 있는 선(모델)을 찾습니다. 
* 모델이 적합되는 과정에서 매번 회귀 계수가 결정이 되고,
* 이때 실제값과 모델 사이의 오류값의 합이 최소가 될 수 있는 최적의 회귀 계수를 찾습니다. (누가? 모델이 자동으로)
* 오류값의 합을 RSS(residual sum of squares)라하고 **비용함수(cost function)** 또는 **손실함수(Loss function)** 라고 합니다. 
* 손실함수 즉, 모델이 얼마나 엉터리인지(?)에 대한 값입니다. 
* 손실함수가 점진적으로 작아지도록 수렴해 가는 과정을 **경사하강법**이란 기법을 사용하여 모델에 구현되어 있다.

## 경사하강법(Gradient Descent)
(Gradient)점진적으로 하강한다.
* 점진적으로 반복적인 계산을 통해 W 파라미터 값을 업데이트하면서 오류값이 최소가 되는 W 파라미터값을 구하는 방식입니다. 

* 최초 오류값이 100었다면, 두번째 오류값은 100보다 작은 90, 세번째는 80으로 지속적으로 오류를 감소시키는 방향으로 W값 업데이트합니다. 
* 그럼 어떻게 오류를 보정할까?
* 편미분 결과값을 반복적으로 보정하면서 w1,w0값을 업데이트합니다.

## L1,L2 정규화(Regularization)
* 다항회귀는 지나치게 모든 데이터에 적합한 회귀식을 만들기 위해서 다항식이 복잡해지고 회귀 계수가 매우 크게 설정이 되면서 과대적합이 됩니다. 
* 회귀 모델은 적절히 데이터에 적합하면서도 회귀 계수가 기하급수적으로 커지는 것을 제어할 수 있어야 합니다. 
* alpha가 0(또는 매우 작은 값)이라면 비용 함수 식은 기존과 동일한 Min(RSS(W) + 0)이 될 것입니다. 
* alpha가 무한대(또는 매우 큰 값)이라면 비용 함수 식은 RSS(W)에 비해 $alpha * ||W||^2$ 값이 너무 커지게 되므로 W값을 0으로 만들어야 Cost가 최소화되는 비용함수 목표를 달성할 수 있습니다. 
* 즉, alpha값을 크게 하면 비용 함수는 회귀 계수 W의 값을 작게 해 과적합을 개선할 수 있으며,
* alpha값을 작게 하면 회귀 계수 W의 값이 커져도 어느 정도 상쇄가 가능합니다. 

image.png

* 비용함수에 alpha값으로 패널티를 부여해 회귀 계수 값의 크기를 감소시켜 과적합을 개선하는 방식을 규제(Regularization)라고 합니다. 
* 규제는 크게 L2방식과 L1방식으로 구분합니다. 
* L2규제는  $alpha * ||W||^2$ 와 같이 W의 제곱에 대해 페널티를 부여하는 방식입니다.
* L2규제를 적용한 회귀를 릿지(Ridge)회귀라고 합니다. 


* L1규제를  $alpha * ||W||$ 와 같이 W의 절대값에 대해 페널티를 부여하는 방식입니다.
* L1규제를 적용한 회귀를 라쏘(Lasso)회귀라고 합니다. 


* ElasticNet은 L2,L1 규제를 함께 결합한 모델입니다.

## 회귀에 사용되는 데이터 형태

* 회귀 모델과 같은 선형 모델은 **피처와 레이블값 간에 선형관계**가 있다고 가정하고, 이러한 최적의 선형함수를 찾아내 결과값을 예측합니다. 
* **피처값과 레이블값의 분포가 정규 분포**(평균을 중심으로 종 모양으로 데이터 값이 분포된 형태)형태를 매우 선호합니다. 
    * 레이블은 반드시 정규분포를 가져야 함, 아니면 로그변환 적용
    * 피처는
        * StandardScaler()클래스 사용: 평균이 0, 분산이 1인 표준 정규 분포를 가지도록 변환
        * MinMaxScaler()클래스 사용: 최소값 0, 최대값 1로 정규화 수행
        * 로그 변환


image.png

* 각 **피처간에는 독립**을 가정합니다. 
* 피처간의 상관관계가 매우 높은 경우 분산이 매우 커져서 오류에 매우 민감해 집니다. 
* 이러한 현상을 다중공선성문제라고 합니다. 
* 대표 피처만 남기고 제거하거나 규제합니다.
